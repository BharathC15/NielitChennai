```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from gapminder import gapminder
df=gapminder.copy()
df.tail()
```

```python
df1=df[df.year==2007]
df1.tail()
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(x="gdpPercap",y="lifeExp",alpha=0.4,data=df1,hue="continent")
plt.xscale("log")
plt.title("Year = 2007")
plt.show()
```

```python
"""
X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
X_scaled = X_std * (max - min) + min
"""
def minmax(x):
    mm=[]
    for i in list(x):
        mm.append(((i-np.min(x))/(np.max(x)-np.min(x)))*10000)
    return pd.Series(mm)
```

```python
minmax(df1["pop"])[0:10]
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(x="gdpPercap",y="lifeExp",alpha=1,data=df1,hue="continent",size=minmax(df1["pop"]))
plt.xscale("log")
plt.title("Year = 2007")
plt.show()
```
