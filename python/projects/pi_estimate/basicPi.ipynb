```python
n=2
bha_pi=0
accuracy=2
acc,num=np.array([]),np.array([])
while True:
    count=0
    for _ in np.arange(n):
        if insideCircle():
            count+=1
    bha_pi=np.round(4*count/n,accuracy)
    #print(bha_pi)
    if bha_pi == np.round(np.pi,accuracy):
        acc=np.append(acc,accuracy)
        num=np.append(num,n)
        break
    if n>100:  #Fail safe Mechanism
        acc=np.append(acc,0)
        num=np.append(num,0)
        break
    n+=1
print("Number of Trials to get accuracy upto {} decimal places is {} ".format(accuracy,n))
```

## Number for trials to get till accuracy of 5 decimal place

```python
#Functions
def bha_rand():
    r = np.random.randint(-100,100)/100
    return r

def insideCircle():
    return np.sqrt((bha_rand())**2+(bha_rand())**2) < 1
```

```python
n=2
bha_pi=0
acc,num=np.array([]),np.array([])

for accuracy in np.arange(1,5):
    while True:
        count=0
        for _ in np.arange(n):
            if insideCircle():
                count+=1
        bha_pi=np.round(4*count/n,accuracy)
        #print(bha_pi)
        if bha_pi == np.round(np.pi,accuracy):
            acc=np.append(acc,accuracy)
            num=np.append(num,n)
            break
        if n>10_000:
            acc=np.append(acc,0)
            num=np.append(num,0)
            break
        n+=1
    print("Number of Trials to get accuracy upto {} decimal places is {} ".format(accuracy,n))
```

```python
plt.figure(dpi=120)
plt.bar(acc,num)
plt.xlabel("Decimal Point Accuracy")
plt.ylabel("No to trials taken")
plt.yscale("log")
plt.show()
```
