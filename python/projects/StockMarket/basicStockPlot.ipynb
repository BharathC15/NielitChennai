```python
import numpy as np
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt
#plt.style.use("seaborn-whitegrid")
```

```python
nifty = pd.read_csv("nifty.csv",index_col=0,parse_dates=True)
nifty.index.name = "Date"
nifty.tail()
```

```python
mpf.plot(nifty.tail(30))
```

```python
mpf.plot(nifty.tail(30),type="candle")
```

```python
mpf.plot(nifty.tail(30),type='candle',mav=(7,12))
```

```python
mpf.plot(nifty.tail(30),type='candle',mav=(7,12),volume=True)
```

```python
mpf.available_styles()
```

```python
mpf.plot(nifty.tail(30),type='candle',volume=True,
         style="yahoo",title="Nifty 50",figscale=1.5)
```

```python
mpf.plot(nifty.tail(30),type='candle',volume=True,
         style="yahoo",title="Nifty 50",figratio=(15,7))
```

```python
mpf.plot(nifty.tail(30),type='candle',mav=(7,12),volume=True,
         style="yahoo",title="Nifty 50",figratio=(15,7))
```
