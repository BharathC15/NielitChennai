```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn-whitegrid")
```

```python
data = pd.read_csv("houseSales.csv")
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
data.columns
```

```python
data.corr()
```

```python
data.corr().index!="price"
```

```python
corr_data = data.corr().loc[data.corr().index!="price","price"]
corr_data
```

```python
np.abs(corr_data)
```

```python
threshold=0.3
corr_data[np.abs(corr_data)>threshold]
```

```python
feature_name = corr_data[np.abs(corr_data)>threshold].index.values
feature_name
```

```python
X = data[feature_name]
print(X.shape)
print(X.head())
```

```python
y = data["price"]
print(y.shape)
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(X_test,y_test)
```
