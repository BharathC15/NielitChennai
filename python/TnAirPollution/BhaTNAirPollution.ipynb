```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_df=pd.read_excel("airQuality.xlsx")
raw_df.head()
```

```python
df=raw_df.iloc[:,[1,3,6,7,8,9,10]].copy()
df.head()
```

```python
df.info()
```

```python
print("Number of Records :",len(df))
print("Percentage of Null Values")
df.isnull().sum()*100/len(df)
```

```python
df.drop(["SPM"],axis=1,inplace=True)
df.head()
```

```python
print("Number of Records :",len(df))
print("Percentage of Null Values")
df.isnull().sum()*100/len(df)
```

```python
plt.figure(dpi=120)
sns.heatmap(df.groupby(["Type of Location","City/Town/Village/Area"]).mean())
```

```python
df.fillna(df.mean(),inplace=True)
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(1, 100))
```

```python
scaler.fit_transform(df.iloc[:,3:])
```

```python
print(np.mean(scaler.fit_transform(df.iloc[:,3:]),axis=1))
print("Number of Records :",len(np.mean(scaler.fit_transform(df.iloc[:,3:]),axis=1)))
```

```python
df["pollution"]=np.mean(scaler.fit_transform(df.iloc[:,3:]),axis=1)
df.head()
```

```python
df.drop(["SO2","NO2","RSPM/PM10"],axis=1,inplace=True)
df.head()
```

```python
df.groupby("City/Town/Village/Area").mean()
```

```python
df["year"]=pd.DatetimeIndex(df["Sampling Date"]).year
df.drop("Sampling Date",axis=1,inplace=True)
df.head()
```

```python
df.groupby(["year","City/Town/Village/Area"]).mean()
```

```python
df1=df.groupby(["year","City/Town/Village/Area"]).mean().reset_index()
df1.head()
```

```python
plt.figure(dpi=120)
sns.scatterplot(x="year",hue="City/Town/Village/Area",legend="full",
                y="pollution",data=df1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
print("Lowest Pollution\n",df1.sort_values(by="pollution").head(1))
print("Highest Pollution\n",df1.sort_values(by="pollution").tail(1))
```

```python
plt.figure(dpi=120)
sns.barplot(x="year",y="pollution",hue="City/Town/Village/Area",data=df1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.figure(dpi=120)
sns.boxplot(x="year",y="pollution",data=df1)
plt.show()
```
