# IRIS dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv("iris.csv")
df.tail()
```

```python
len(df)
```

```python
df.species.values
```

```python
df["species"].value_counts()
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df,hue="species",diag_kind="hist")
```

```python
X=df.iloc[:,:4]
X.head()
```

```python
from sklearn.decomposition import PCA
```

```python
bha=PCA(n_components=2)
```

```python
bha1=bha.fit_transform(X)
```

```python
bha1
```

```python
bha2=pd.DataFrame(bha1,columns=["x1","x2"])
print(len(bha2))
bha2.head()
```

```python
sns.scatterplot(x=bha2.x1,y=bha2.x2)
```

```python
sns.scatterplot(x=bha2.x1,y=bha2.x2,hue=df.species)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
km.fit(X)
```

```python
y_cluster_kmeans = km.predict(X)
y_cluster_kmeans
```

```python
sns.scatterplot(x=bha2.x1,y=bha2.x2,hue=y_cluster_kmeans)
```

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
sns.scatterplot(x=bha2.x1,y=bha2.x2,hue=df.species)
plt.subplot(1,2,2)
sns.scatterplot(x=bha2.x1,y=bha2.x2,hue=y_cluster_kmeans)
plt.show()
```
