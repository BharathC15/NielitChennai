# Poisson Distribution 
### English pronunciation:  /ˈpwɑːsɒn/

```python
# Importing the essential Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from scipy.stats import poisson
```

* Expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event. 
* The Poisson distribution can also be used for the number of events in other specified intervals such as distance, area or volume.

```python
data=poisson.rvs(mu=5.5,size=100)
data
```

```python
np.unique(data,return_counts=True)
```

```python
plt.figure(dpi=120)
sns.distplot(data)
plt.xlabel("Random Variable")
plt.ylabel("Probability")
plt.show()
```

```python
poisson.pmf(k=2,mu=5.5)
```

```python
for i in np.arange(0,15):
    print(i,"->",poisson.pmf(k=i,mu=5.5))
```

```python
rv,prob=[],[]
for i in np.arange(0,15):
    rv.append(i)
    prob.append(poisson.pmf(k=i,mu=5.5))
df=pd.DataFrame({"rv":rv,"Probability":prob})
plt.figure(dpi=120)
sns.barplot(x="rv",y="Probability",data=df)
plt.show()
```

```python
rv,pdf_data,cdf_data=[],[],[]
for i in np.arange(0,15):
    rv.append(i)
    pdf_data.append(poisson.pmf(k=i,mu=5.5))
    cdf_data.append(poisson.cdf(k=i,mu=5.5))
df=pd.DataFrame({"rv":rv,"PDF":pdf_data,"CDF":cdf_data})
df
```

```python
plt.figure(dpi=120)
plt.plot(df["rv"],df["PDF"],label="PDF")
plt.plot(df["rv"],df["CDF"],label="CDF")
plt.xlabel("Random Variable")
plt.legend()
plt.show()
```

```python
data1=poisson.rvs(mu=0.1,size=10000)
data2=poisson.rvs(mu=0.5,size=10000)
data3=poisson.rvs(mu=1,size=10000)
data4=poisson.rvs(mu=1.5,size=10000)
data5=poisson.rvs(mu=2,size=10000)
data6=poisson.rvs(mu=3,size=10000)
data7=poisson.rvs(mu=5,size=10000)
data8=poisson.rvs(mu=7,size=10000)
data9=poisson.rvs(mu=10,size=10000)
data10=poisson.rvs(mu=20,size=10000)
data11=poisson.rvs(mu=50,size=10000)

data1_variable,data1_counts=np.unique(data1,return_counts=True)
data2_variable,data2_counts=np.unique(data2,return_counts=True)
data3_variable,data3_counts=np.unique(data3,return_counts=True)
data4_variable,data4_counts=np.unique(data4,return_counts=True)
data5_variable,data5_counts=np.unique(data5,return_counts=True)
data6_variable,data6_counts=np.unique(data6,return_counts=True)
data7_variable,data7_counts=np.unique(data7,return_counts=True)
data8_variable,data8_counts=np.unique(data8,return_counts=True)
data9_variable,data9_counts=np.unique(data9,return_counts=True)
data10_variable,data10_counts=np.unique(data10,return_counts=True)
data11_variable,data11_counts=np.unique(data11,return_counts=True)
```

```python
plt.figure(dpi=120)
plt.plot(data1_variable,data1_counts/10000,label="mu=0.1",linewidth=3,alpha=0.7)
plt.plot(data3_variable,data3_counts/10000,label="mu=1",linewidth=3,alpha=0.7)
plt.plot(data6_variable,data6_counts/10000,label="mu=3",linewidth=3,alpha=0.7)
plt.plot(data7_variable,data7_counts/10000,label="mu=5",linewidth=3,alpha=0.7)
plt.plot(data9_variable,data9_counts/10000,label="mu=10",linewidth=3,alpha=0.7)
plt.xlabel("Random Variable")
plt.ylabel("Probability")
plt.title("Poisson Distribution for different mu values")
plt.legend()
plt.show()
```

```python
def poisson_gen(mu):
    rv,pdf_data,cdf_data=[],[],[]
    for i in np.arange(0,15):
        rv.append(i)
        pdf_data.append(poisson.pmf(k=i,mu=mu))
        cdf_data.append(poisson.cdf(k=i,mu=mu))
    df=pd.DataFrame({"rv":rv,"PDF":pdf_data,"CDF":cdf_data})
    return df
```

```python
data=poisson_gen(5.5)
data.head()
```

```python
plt.figure(dpi=120)
x=[0.1,1,3,5,10]
for i in x:
    plt.plot(poisson_gen(i).rv,poisson_gen(i).PDF,label="mu = "+str(i))
plt.xlabel("Random Variable")
plt.ylabel("Probability")
plt.title("Poisson Distribution for different mu values")
plt.legend()
plt.show()
```
