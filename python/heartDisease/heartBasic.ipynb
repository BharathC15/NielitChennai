```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(dpi=120,figsize=(15,10))
sns.heatmap(df.corr(),center=0,cmap="BrBG",vmin=-1,vmax=1,annot = True,
           cbar_kws= {'orientation': 'horizontal'})
plt.title("Correlation Heat Map")
plt.show()
```

```python
pd.DataFrame(np.tril(df.corr()))
```

```python
np.triu(df.corr()).shape
```

```python
plt.figure(dpi=120,figsize=(15,10))
sns.heatmap(df.corr(),center=0,cmap="BrBG",vmin=-1,vmax=1,annot = True,
           cbar_kws= {'orientation': 'horizontal'},mask=np.triu(df.corr()))
plt.title("Correlation Heat Map")
plt.show()
```

```python
# Automatically Extracting the features based on a Threshold value
threshold=float(input("Enter a threshold value between 0.0 to 1.0 :"))
df1=df.corr().iloc[-1,:-1].copy()
print("DataFrame with the columns \nHaving Correlation less than",-threshold,"and greater than",threshold)
df[df1[np.absolute(df1)>=threshold].index]
```
